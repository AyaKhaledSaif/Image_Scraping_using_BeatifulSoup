```python
import requests
from bs4 import BeautifulSoup
import shutil
```

```python
link = 'https://eg.hatla2ee.com/en/car/peugeot'
res = requests.get(link)
soup = BeautifulSoup(res.text,'html.parser')
```

```python
divs = soup.find_all('div', attrs = {'class':'newCarListUnit_header'})
```

```python
carDivsLink = []
img_links = []
count = 0
for div in range(len(divs)):
    link1 = 'https://eg.hatla2ee.com'+ divs[div].find('a').get('href')
    carDivsLink.append(link1)

for carLink in carDivsLink:
    counter = 0
    res = requests.get(carLink)
    soup = BeautifulSoup(res.text,'html.parser')
    swip_lazy = soup.find_all('img',attrs={'class':'swiper-lazy'})
    print('num of car images: ',len(swip_lazy))
    #print(swip_lazy, '\n','-'*60)
    for imgTag in swip_lazy:
        imgUrl = imgTag.get('data-src')
        imgUrl_name = imgTag.get('alt')
        counter += 1
        print(counter,': ',imgUrl)
        
        if len(swip_lazy) == 0:
            print('\n Image Couldn\'t be retrieved')
        else:
            count += 1
            res = requests.get(imgUrl, stream = True)
            if res.status_code == 200:
                with open(imgUrl_name+str(count)+'.jpg','wb') as f:
                    shutil.copyfileobj(res.raw, f)
                print('\n Image sucessfully Downloaded: ',imgUrl_name)
          
    
        
    

```
